## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Research question:** Given a page's observable signals (impressions, position, CTR, staleness, word count), can I rank pages so an editor's top-N review queue is disproportionately pages that are actually declining — better than a hand-written rule?

**Unit of analysis:** One content item (one row in content_refresh_anonymized.csv).

**Decision this improves:** Which pages an editor reviews for refresh this week, out of thousands.

**Action taken:** An editor opens the top-N pages from the ranked queue and reviews them for refresh, instead of picking pages at random.

**Cost of a wrong call:** Flagging a page that isn't really declining wastes editor time (low, recoverable cost). Missing a genuinely declining page (false negative) lets it keep losing traffic silently until the next audit (slower, compounding cost). So precision at the top of the queue matters more than perfect recall — the editor's time is the scarce resource.

**Why data/ML helps at all:** A fixed rule (e.g. "stale AND high impressions") is transparent but can't weigh several weak signals together. My own pipeline run already shows a learned model roughly tripling precision over the hand-written rule at the top of the queue — real evidence there's learnable structure a simple if-statement misses.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Numbers below come straight from the starter CSV and from my own committed pipeline output (outputs/model_report.md), not memory.

```python
import os

if not os.path.exists("flyrank-ml-track"):
    !git clone https://github.com/muneeb-khokhar/flyrank-ml-track.git

os.chdir("flyrank-ml-track")

import pandas as pd, re

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

decline_rate = df["trend_direction"].eq("down").mean()
stale_share = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).mean()

print(f"Rows: {len(df):,}")
print(f"Declining-label rate (trend_direction == 'down'): {decline_rate:.1%}")
print(f"Share flagged 'stale_visible_page' by the baseline rule: {stale_share:.1%}")

report = open("outputs/model_report.md").read()
baseline_p50 = float(re.search(r"baseline_rules \| [\d.]+ \| [\d.]+ \| ([\d.]+)", report).group(1))
rf_p50 = float(re.search(r"random_forest \| [\d.]+ \| [\d.]+ \| ([\d.]+)", report).group(1))
print(f"Baseline rule  Precision@50: {baseline_p50:.3f}")
print(f"Random forest  Precision@50: {rf_p50:.3f}  (~{rf_p50/baseline_p50:.1f}x the baseline)")
```

I'm not predicting Google's algorithm. trend_pct/trend_direction are never features — they define the label itself, so feeding them in would just let the model copy the answer (this is the "label trap" both skills/flyrank/flyrank-data/SKILL.md and the lane guide warn about).
